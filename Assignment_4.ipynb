### Copy.copy() or shallow copy

```python
import copy
lst1=[[1,2,3,4],[5,6,7,8]]
lst2=copy.copy(lst1)
```

```python
lst1
```

```python
lst2
```

```python
lst2[1][0]=100
```

```python
lst2
```

```python
lst1
```

```python
# lst1 and lst2 in both of them values are changed 
```

### Deepcopy()

```python
import copy
lst1=[[1,2,3,4],[5,6,7,8]]
lst2=copy.deepcopy(lst1)
```

```python
lst1
```

```python
lst2
```

```python
lst2[1][0]=100
```

```python
lst2
```

```python
lst1
```

```python
# from above cases it can be seen that in case of deepcopy
# values are changed in lst2 only not in lst1

# but in case of shallow copy values were changed in both 
# lst1 and lst 2

# this is the main diff b/w copy.copy() and copy.deepcopy()
```
